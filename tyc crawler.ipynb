```python
import time
from selenium import webdriver
from selenium.webdriver.common.by import By
import pandas as pd
from bs4 import BeautifulSoup
import warnings
warnings.filterwarnings('ignore')
```

```python
option = webdriver.ChromeOptions()
option.add_argument('user-agent=Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/119.0.0.0 Safari/537.36 Edg/119.0.0.0')
driver = webdriver.Chrome(options=option)
driver.implicitly_wait(5)
driver.get('https://www.tianyancha.com/')
driver.maximize_window()
```

```python
def get_result_table(result_html):
    soup = BeautifulSoup(result_html,"lxml")
    company_name = soup.find("h1",class_="index_company-name__LqKlo").text
    x = str(soup.find("table","index_tableBox__ZadJW"))
    result_table = pd.read_html(x)[0]
    book_value = result_table[result_table[2]=='注册资本'][3].values[0]
    company_field = result_table[result_table[2]=='行业'][3].values[0]
    #company_address = result_table[result_table[0]=='注册地址'][1].values[0]
    company_area = result_table[result_table[0]=='经营范围'][3].values[0]
    result = pd.DataFrame([[company_name,book_value,company_field,company_area]])
    return result
```

```python
final_result = pd.DataFrame()
file_list = list(pd.read_excel('名单.xlsx')['name'])
driver.switch_to.window(driver.window_handles[-1])
text_label = driver.find_element_by_xpath('//*[@id="page-container"]/div[1]/div/div[3]/div[2]/div[1]/div[1]/input')
text_label.clear()
text_label.send_keys("阿里巴巴")
time.sleep(1)
button = driver.find_element_by_xpath('//*[@id="page-container"]/div[1]/div/div[3]/div[2]/div[1]/button')
button.click()
driver.implicitly_wait(5)

for file in file_list[0:]:
    driver.switch_to.window(driver.window_handles[-1])
    #time.sleep(5)
    driver.find_element_by_xpath('//*[@id="page-header"]/div/div[2]/div/div/div/span/i').click()
    driver.find_element_by_xpath('//*[@id="page-header"]/div/div[2]/div/div/div[1]/input').send_keys(file)
    time.sleep(1)
    button = driver.find_element_by_xpath('//*[@id="page-header"]/div/div[2]/div/div/button')
    button.click()
    try:
        driver.find_element_by_xpath('//*[@id="page-container"]/div/div[2]/section/main/div[1]/div/div/div[3]/div/span').click()
        #省份展开按钮
        time.sleep(1)
        driver.find_element_by_xpath('//*[@id="page-container"]/div/div[2]/section/main/div[1]/div/div/div[2]/div/div[1]/div/div[2]').click()
        #省份更多按钮
        time.sleep(1)
        driver.find_element_by_xpath('//*[@id="page-container"]/div/div[2]/section/main/div[1]/div/div/div[2]/div/div[1]/div/div[1]/div/div[11]/span').click()
        #省份选择按钮
        time.sleep(1)
        driver.find_element_by_xpath('//*[@id="page-container"]/div/div[2]/section/main/div[1]/div/div/div[2]/div[2]/div[1]/div[2]/div/div/div[2]/span').click()
        #城市选择按钮
        time.sleep(1)
        driver.implicitly_wait(5)
        try:
            driver.find_element_by_xpath('//*[@id="page-container"]/div/div[2]/section/main/div[2]/div[2]/div[1]/div/div[2]/div[2]/div[1]/div[1]/a/span').click()
            driver.switch_to.window(driver.window_handles[-1])
            try:
                result_table = get_result_table(driver.page_source)
                final_result = final_result._append(result_table)
                driver.close()
            except:
                final_result = final_result._append(pd.Series(), ignore_index=True)
                driver.close()
        except:
            final_result = final_result._append(pd.Series(), ignore_index=True)
    except:
        driver.find_element_by_xpath('//*[@id="page-container"]/div/div[2]/section/main/div[2]/div/div/div[3]/div/span').click()
        #省份展开按钮
        time.sleep(1)
        driver.find_element_by_xpath('//*[@id="page-container"]/div/div[2]/section/main/div[2]/div/div/div[2]/div/div[1]/div/div[2]').click()
        #省份更多按钮
        time.sleep(1)
        driver.find_element_by_xpath('//*[@id="page-container"]/div/div[2]/section/main/div[2]/div/div/div[2]/div/div[1]/div/div[1]/div/div[11]/span').click()
        #省份选择按钮
        time.sleep(1)
        driver.find_element_by_xpath('//*[@id="page-container"]/div/div[2]/section/main/div[1]/div/div/div[2]/div[2]/div[1]/div[2]/div/div/div[2]/span').click()
        #城市选择按钮
        time.sleep(1)
        driver.implicitly_wait(5)
        try:
            driver.find_element_by_xpath('//*[@id="page-container"]/div/div[2]/section/main/div[1]/div/div/div[2]/div[2]/div[1]/div[2]/div/div/div[1]/span').click()
            driver.switch_to.window(driver.window_handles[-1])
            try:
                result_table = get_result_table(driver.page_source)
                final_result = final_result._append(result_table)
                driver.close()
            except:
                final_result = final_result._append(pd.Series(), ignore_index=True)
                driver.close()
        except:
            final_result = final_result._append(pd.Series(), ignore_index=True)
```
